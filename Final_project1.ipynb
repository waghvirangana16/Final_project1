```python
class Restaurent:
    def __init__(self):
        self.food={}
        self.food_id=len(self.food)+1
    def food_items(self):
        self.name=input("Enter a name: ")
        self.quantity=int(input("Enter a quantity: "))
        self.price=int(input("Enter a price: $" ))
        self.discount=int(input("Enter a discount: $"))
        self.stock=int(input("Enter available stock "))
        self.list={"name":self.name,"quantity":self.quantity,"price":self.price,"discount":self.discount,"stock":self.stock}
        self.food[self.food_id]=self.list
        self.food_id=len(self.food)+1
        print(self.food)
    def edit_item(self):
        e=int(input("Enter a food id which you want to edit "))
        for i in self.food[e]:
            self.food[e][i]=input(f"Enter the {i} that you want to update ")
        print("\n",self.food)
    def view_item(self):
        print("List of all food items is ")
        for k in self.food:
            print("food_id :",k)
            for j in self.food[k]:
                print(j,":",self.food[k][j])
            print("\n****")
    def remove_item(self):
        d=int(input("Enter a food id which you want to delete "))
        del self.food[d]
        print("Final list is: ",self.food)
    
x=Restaurent()
x=Restaurent()
x=Restaurent()
x.food_items()
print("\n****************")
x.food_items()
print("\n****************")
x.food_items()
print("\n****************")
x.edit_item()
print("\n****************")
x.view_item()
x.remove_item()

```
